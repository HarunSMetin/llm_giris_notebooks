```python
from dotenv import load_dotenv
import os

# .env dosyasından ortam değişkenlerini yükleyin API erişimi için
load_dotenv()
```

### Langchain
Langchain, LLM uygulamaları oluşturmak için kapsamlı bir SDK'dır ve model sağlayıcıları arasında kolayca geçiş yapılabilmesini sağlar. LLM API'lerini çağırma, çıktıları yönetme, ve çeşitli dil modelleriyle entegrasyon yapma sürecini basitleştirir.

### Büyük Dil Modeli
OpenAI'nin GPT-4 gibi Büyük Dil Modelleri (LLM'ler), insan benzeri metni anlamak ve oluşturmak için eğitilmiştir. Birçok GenAI uygulamasının temelini oluştururlar ve geniş bir doğal dil işleme görevine olanak tanırlar.

```python
from langchain_openai import ChatOpenAI
# Modelimizi GPT-4 ile başlatın
model = ChatOpenAI(model="gpt-4o")
```

### LangSmith
LangSmith, LLM uygulamalarını daha iyi anlamak ve hata ayıklamak için bir araçtır. LLM ile doğrudan etkileşim kurulmasını, ek açıklama kuyrukları ve test veri setleri oluşturulmasını kolaylaştırır ve model performansını iyileştirmeye yardımcı olur.

### İlk Prompt'unuzu Oluşturmak
LLM'ler farklı türlerde mesajları işler. SystemMessage, LLM'nin rolünü ve işlevini tanımlar ve kullanıcı girdilerini (HumanMessage) nasıl işleyeceğini yönlendirir.

```python
from langchain_core.messages import HumanMessage, SystemMessage

# Etkileşim için sistem ve insan mesajlarını tanımlayın
system_message = "Kullanıcının selamını İspanyolca olarak karşılan ve kendinizi tanıtın."
human_message = "Merhaba AI! Benim adım HARUN"

messages = [
    SystemMessage(content=system_message),
    HumanMessage(content=human_message),
]

# Tanımlanan mesajlarla LLM API'sini çağırın
model.invoke(messages)
```

### Çıktı Biçimlendirmesi
LangChain'in çıktı ayrıştırıcıları, LLM çıktılarını biçimlendirme işlemini basitleştirir ve ek meta veri olmadan ana metin yanıtının kolayca çıkarılmasını sağlar.

```python
from langchain_core.output_parsers import StrOutputParser

# Dize çıktı ayrıştırıcısını başlatın
parser = StrOutputParser()
```

```python
# LLM'nin çıktısını biçimlendirmek için ayrıştırıcıyı uygulayın
chain = model | parser
chain.invoke(messages)
```

### Model Değiştirme
LangChain, farklı LLM'ler arasında kolayca geçiş yapılabilmesini sağlayıp belirli görevler için en uygun modeli seçme esnekliği sunar.

```python
from langchain_anthropic import ChatAnthropic
# Karşılaştırma için Anthropic Claude 3'e geçin
model = ChatAnthropic(model="claude-3-opus-20240229")
```

```python
# Yeni LLM'yle zinciri yeniden çalıştırın
chain = model | parser
chain.invoke(messages)
```

#### LangSmith ile İzleri Gözden Geçirme
LangSmith, LLM etkileşimlerinin ayrıntılı izlenmesine olanak tanır, modelin davranışına ilişkin içgörüler sağlar ve hata ayıklama ile iyileştirmeyi kolaylaştırır.
