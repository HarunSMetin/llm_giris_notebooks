## Giriş
Bu notebook, Langchain ve OpenAI'nin GPT-4 LLM'i kullanarak bir URL'den içerik özeti oluşturan bir Dil Öğrenme Modeli (LLM) uygulaması oluşturma sürecinde sizi yönlendirmek için tasarlanmıştır.

#### URL ---> İçerik (docslar) ---> LLM ---> Özet

Bu işlem, belirtilen bir URL'den içerik alınmasını, içerikten ilgili bilgileri çıkarmak için işlenmesini ve ardından kısa bir özet oluşturmak için LLM kullanılmasını içerir. Bu notebook, bu işlemin her adımında sizi yönlendirecek, detaylı açıklamalar ve örnekler sağlayacaktır.

```python
from dotenv import load_dotenv
import os

# .env dosyasından ortam değişkenlerini yükleyin
load_dotenv()
```

#### Belge Yükleyiciler
Belge yükleyiciler, LangChain'in LLM uygulamaları için kullanılabilecek içerik almanın sağlandığı bir yetenektir. İçeriği özetlemeden önce çekmek ve hazırlamak için gereklidir. Bu örnekte, belirtilen bir URL'den içerik yüklemek için WebBaseLoader kullanacağız.

```python
# Bir URL'den içerik yükleyin, tüm yazı aslında okunmaya değer
from langchain_community.document_loaders import WebBaseLoader
url = "https://lilianweng.github.io/posts/2023-06-23-agent/"
loader = WebBaseLoader(url)
docs = loader.load()
```

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4o")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser

system_message = "Lütfen içerik için kısa bir özet sağlayın"
human_message = "İçerik: " + str(docs)

messages = [
    SystemMessage(content=system_message),
    HumanMessage(content=human_message),
]

parser = StrOutputParser()

chain = model | parser
# Sistem ve insan mesajını LLM API'sine geçin ve çağırın
chain.invoke(messages)
```
