## Giriş
Bu notebook, Langchain ve OpenAI'nin GPT-4 çok modlu yeteneği kullanarak bir Büyük Dil Modeli (LLM) uygulaması oluşturma sürecinde sizi yönlendirmek için tasarlanmıştır. Bu model görüntüleri de anlayabilir.

```python
from dotenv import load_dotenv
import os

# .env dosyasından ortam değişkenlerini yükleyin
load_dotenv()
```

```python
image_url = "https://systemsarchitect.io/assets/images/diagram-app-structural-flow-b0cc18c0e4efee46b1ecc0db9f40cde1.png"
```

```python
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-4o")
```

```python
message = HumanMessage(
    content=[
        {"type": "text", "text": "görüntüdeki mimari diyagramını açıkla"},
        {"type": "image_url", "image_url": {"url": image_url}},
    ],
)
response = model.invoke([message])
print(response.content)
```
