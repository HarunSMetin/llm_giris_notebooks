## Giriş
Bu notebook, Langchain ve OpenAI'nin GPT-4 LLM'i kullanarak basit bir sohbet uygulaması oluşturma sürecinde sizi yönlendirmek için tasarlanmıştır. Odak noktası, sorunsuz bir konuşma deneyimi için sohbet geçmişini nasıl yönetileceğini göstermektir.

### Sohbet
Çoğu zaman bir LLM'ye sadece bir soru sormak istemeyiz. Aslında ileri geri bazı etkileşimler olan bir oturumdur. Bu sohbet deneyimini etkinleştirmek için önceki mesajları sohbet geçmişi ile takip etmemiz gerekir. LLM'ler durum bilgisiz olur ve yalnızca kendi bağlam penceresi içindeki bilgilerle uğraşabilir. Yani bu sohbet geçmişini ayrı olarak saklamam ve yeni mesaj ile birlikte LLM'ye geçmemiz gerekir, böylece tüm bağlamı olur. Bu notebook, konuşmalarda bağlamı korumak için sohbet geçmiş yönetiminin nasıl uygulanacağını göstermektedir.

Başlamak için notebook birde yaptığımız merhaba dünya örneğini kurup sohbet geçmişinin gerekliliğini vurgulamak için ayarlayalım

```python
from dotenv import load_dotenv
import os

# .env dosyasından ortam değişkenlerini yükleyin
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
# Sohbet uygulaması için modelimizi GPT-4 ile başlatın
model = ChatOpenAI(model="gpt-4o")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

system_message = "Kullanıcının selamını İspanyolca olarak karşılayın"
human_message = "Merhaba AI! Benim adım GP"

messages = [
    SystemMessage(content=system_message),
    HumanMessage(content=human_message),
]
```

```python
model.invoke(messages)
```

Şimdi önceki sorumuza ne olduğunu sormayalım ve hatırlayıp hatırlamadığını görelim

İpucu: hatırlamayacak çünkü LLM'ler durum bilgisiz

```python
human_message = "Adım ne?"
```

```python
model.invoke(human_message)
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}


def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]


with_message_history = RunnableWithMessageHistory(model, get_session_history)
config = {"configurable": {"session_id": "1"}}
```

```python
with_message_history.invoke(messages, config=config)
```

```python
human_message = "Adım ne?"
with_message_history.invoke(human_message, config=config)
```
